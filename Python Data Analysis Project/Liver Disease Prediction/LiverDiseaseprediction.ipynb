```python
df_young=df[df['Age']<=40]
df_old=df[df['Age']>=60]
df_middle=df[(df['Age']>40)&(df['Age']<60)]
```

```python
yng_diag=df_young['Diagnosis'].value_counts()
mid_diag=df_middle['Diagnosis'].value_counts()
old_diag=df_old['Diagnosis'].value_counts()
print("Young Diag", yng_diag)
print("Mid Diag",mid_diag)
print("Old Diag",old_diag)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
X_data=df.drop(['Diagnosis'],axis=1)
Y_data=df['Diagnosis']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_data,Y_data,test_size=0.2,random_state=2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred)
```
